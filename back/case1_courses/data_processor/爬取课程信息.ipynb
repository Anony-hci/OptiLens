```python
import requests
from bs4 import BeautifulSoup

# 基本的网页URL
BASE_URL = "https://zhjwxk.cic.tsinghua.edu.cn/xkYjs.vxkYjsJxjhBs.do?m=kkxxSearch&p_xnxq=2024-2025-2"  # 这里替换为实际的网址

# 存储所有爬取的数据
courses = []

# 获取页面内容的函数
def get_page_content(page_number):
    # 构造完整的URL
    url = f"{BASE_URL}?page={page_number}"
    response = requests.get(url)
    
    if response.status_code == 200:
        return response.text
    else:
        print(f"Failed to retrieve page {page_number}")
        return None

# 解析页面内容，提取开课信息
def parse_page(content):
    soup = BeautifulSoup(content, 'html.parser')
    
    # 假设每个课程信息在一个class为"course-info"的div标签内
    course_items = soup.find_all('div', class_='course-info')
    
    for course in course_items:
        # 根据实际页面的结构提取你需要的数据
        course_name = course.find('h2').get_text(strip=True)  # 课程名称
        course_teacher = course.find('span', class_='teacher').get_text(strip=True)  # 教师
        course_time = course.find('span', class_='time').get_text(strip=True)  # 上课时间
        
        # 将爬取的课程信息存储
        courses.append({
            'name': course_name,
            'teacher': course_teacher,
            'time': course_time
        })

# 主爬虫逻辑
def crawl_courses():
    # 假设总共有50页课程信息
    total_pages = 250
    
    for page_number in range(1, total_pages + 1):
        print(f"Scraping page {page_number}...")
        
        # 获取页面内容
        content = get_page_content(page_number)
        
        if content:
            # 解析页面内容并提取课程信息
            parse_page(content)
    
    print(f"Scraping completed. Total courses found: {len(courses)}")
    return courses

# 调用爬虫并打印结果
if __name__ == "__main__":
    courses = crawl_courses()
    
    # 输出所有爬取到的课程信息
    for course in courses:
        print(course)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 设置基础URL和页数范围
base_url = "https://zhjwxk.cic.tsinghua.edu.cn/xkYjs.vxkYjsJxjhBs.do?m=kkxxSearch&p_xnxq=2024-2025-2?page="  # 假设这是课程列表的基本URL
num_pages = 10  # 假设共有50页

# 定义一个空的列表来存储所有课程数据
courses = []

# 遍历每一页
for page_num in range(1, num_pages + 1):
    url = base_url + str(page_num)
    response = requests.get(url)
    
    if response.status_code == 200:
        print(str(page_num) + "success")
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # 查找所有课程信息行，假设每个课程信息都在<tr class="trr2">中
        rows = soup.find_all('tr', class_='trr2')
        print(soup)
        
        for row in rows:
            # 提取每一行的各个字段
            course_info = {}
            
            course_info['开课院系'] = row.find_all('td')[0].get_text(strip=True)
            course_info['课程号'] = row.find_all('td')[1].get_text(strip=True)
            course_info['课序号'] = row.find_all('td')[2].get_text(strip=True)
            course_info['课程名'] = row.find_all('td')[3].get_text(strip=True)
            course_info['学分'] = row.find_all('td')[4].get_text(strip=True)
            course_info['主讲教师'] = row.find_all('td')[5].get_text(strip=True)
            course_info['本科生课容量'] = row.find_all('td')[6].get_text(strip=True)
            course_info['本科生课余量'] = row.find_all('td')[7].get_text(strip=True)
            course_info['研究生课容量'] = row.find_all('td')[8].get_text(strip=True)
            course_info['研究生课余量'] = row.find_all('td')[9].get_text(strip=True)
            course_info['上课时间'] = row.find_all('td')[10].get_text(strip=True)
            course_info['课程特色'] = row.find_all('td')[11].get_text(strip=True)
            course_info['年级'] = row.find_all('td')[12].get_text(strip=True)
            course_info['是否二级选课'] = row.find_all('td')[13].get_text(strip=True)
            course_info['实验信息'] = row.find_all('td')[14].get_text(strip=True)
            course_info['重修是否占容量'] = row.find_all('td')[15].get_text(strip=True)
            course_info['是否选课时限制'] = row.find_all('td')[16].get_text(strip=True)
            course_info['通识选修课组'] = row.find_all('td')[17].get_text(strip=True)
            
            # 将数据添加到列表中
            courses.append(course_info)
    
    else:
        print(f"Failed to retrieve page {page_num}")

# 将数据转换为DataFrame
df = pd.DataFrame(courses)

# 输出数据到CSV文件
df.to_csv('courses.csv', index=False, encoding='utf-8')

print("Data extraction complete. Data saved to 'courses.csv'.")
```
