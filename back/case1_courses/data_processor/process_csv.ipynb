```python
import random
import pandas as pd

# 读取课程数据
df = pd.read_csv('../../front/src/data/courses.csv')

def get_dropout_rate(row):
    # 获取课程名和余量信息
    course_name = row['课程名']
    remaining = int(row['本科生课余量'])
    capacity = int(row['本科生课容量'])
    
    # 如果没有课余量
    if remaining == 0:
        # 基础掉课率为30-50%
        base_rate = random.uniform(30, 50)
        
        # 检查课程名中是否包含特定关键词
        keywords = ['人工智能', 'AI', '机器学习', '深度学习', '数据科学', '计算机']
        if any(keyword in course_name for keyword in keywords):
            # 对于热门课程,提高掉课率到80-97%
            return random.uniform(80, 97)
        else:
            # 其他无余量课程返回基础掉课率
            return base_rate
    else:
        # 还有余量的课程,掉课率较低(10-20%)
        return random.uniform(10, 20)

# 应用函数生成掉课率
df['掉课率'] = df.apply(get_dropout_rate, axis=1)

# 将掉课率四舍五入到2位小数
df['掉课率'] = df['掉课率'].round(2)
# 只保留指定列
df = df[['课程名', '学分', '主讲教师', '上课时间', '开课院系', '掉课率']]

# 导出为CSV文件
df.to_csv('../../front/src/data/courses_with_dropout.csv', index=False, encoding='utf-8')

```

```python
import random
import pandas as pd
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.path.dirname('__file__'), '..')))
from llm.chat import openai_client, construct_user


class fixTimePrompt:
    def __init__(self, time_str):
        self.SystemPrompt = '''
这个是我收集上来的课程数据，其中的上课时间因为屏幕大小会存在一些缺失，即存在'...'
请帮我fix 这个问题，时间格式包括以下几种类型：
3-2(前八周)
2-6(后八周)
2-6(全周)
1-4(后八周),1-3(后八周)
2-4(单周),2-3(单周)
5-4(双周)
3-4(1-15周),3-3(1-15周)
1-6(5周),3-6(1-2,4-7周)
3-6(1-2,4-7周),3-6(3周)
5-1(1,3-14周)
'''
        self.UserPrompt = f'''
上课时间为: {time_str},
请直接返回一个可能的上课时间，不要解释, 不要添加"上课时间为: "前缀。我会将结果直接插入表格。
'''




def fixTime(time_str):
    prompt = fixTimePrompt(time_str)
    response = openai_client.sync_call_gpt(prompt.SystemPrompt, [construct_user(prompt.UserPrompt)], "gpt-4o")
    print(response)
    return response

# 读取课程数据
df = pd.read_csv('../../front/src/data/courses_with_dropout.csv')

# 检查并修复上课时间格式
df['上课时间'] = df['上课时间'].apply(lambda x: fixTime(x) if '…' in x else x)


# 导出为CSV文件
df.to_csv('../../front/src/data/courses_fixed_time.csv', index=False, encoding='utf-8')
```
